# Load the Electronics Dataset

```python
import pandas as pd
import json

def process_batch(batch, batch_index):
    df_batch = {}
    for line_index, line in enumerate(batch):
        try:
            data = json.loads(line.strip())
            df_batch[line_index] = data
        except Exception as e:
            print("Error:", e)
            print("Problematic line in batch {}: {}, index {}".format(batch_index, line, line_index))
    return pd.DataFrame.from_dict(df_batch, orient='index')

def getDF(path, max_batch_size=10000):
    with open(path, 'r', encoding='utf-8') as f:
        batch_index = 0
        total_rows = 6739590  # Total number of rows in your dataset
        while batch_index * max_batch_size < total_rows:
            remaining_rows = total_rows - batch_index * max_batch_size
            batch_size = min(max_batch_size, remaining_rows)
            batch = [next(f) for _ in range(batch_size)]
            print("Processing batch:", batch_index)
            yield process_batch(batch, batch_index)
            batch_index += 1
            pd.DataFrame().pipe(lambda x: x)
        print("End of file reached.")

# Initialize an empty list to store the DataFrames of each batch
dfs = []

# Iterate over the batches returned by the generator
for batch_df in getDF('Electronics_5.json'):
    dfs.append(batch_df)

# Concatenate all DataFrames into a single DataFrame
df = pd.concat(dfs, ignore_index=True)

# Display the first few rows of the DataFrame
print(df.head())
```

# Loading the Meta-Data

```python
import pandas as pd
import json

def process_batch(batch, batch_index):
    df_batch = {}
    for line_index, line in enumerate(batch):
        if line is not None:
            try:
                data = json.loads(line.strip())
                df_batch[line_index] = data
            except Exception as e:
                print("Error:", e)
                print("Problematic line in batch {}: {}, index {}".format(batch_index, line, line_index))
    return pd.DataFrame.from_dict(df_batch, orient='index')

def getDF(path, max_batch_size=10000):
    with open(path, 'r', encoding='utf-8') as f:
        batch_index = 0
        total_rows = 786868   # Total number of rows in your dataset
        while True:
            batch = [next(f, None) for _ in range(max_batch_size)]
            if not any(batch):
                break
            print("Processing batch:", batch_index)
            yield process_batch(batch, batch_index)
            batch_index += 1
            pd.DataFrame().pipe(lambda x: x)
        print("End of file reached.")

# Initialize an empty list to store the DataFrames of each batch
dfs = []

# Iterate over the batches returned by the generator
for batch_df in getDF('meta_Electronics.json'):
    dfs.append(batch_df)

# Concatenate all DataFrames into a single DataFrame
df_meta = pd.concat(dfs, ignore_index=True)

# Display the first few rows of the DataFrame
print(df_meta.head())
```
